# Titanic Survival Prediction

## Project Overview

The Titanic dataset contains information about passengers who were aboard the Titanic when it sank on April 15, 1912.

The objective of this project is to analyze passenger information and build a machine learning model that predicts whether a passenger survived or not.

This is a **binary classification problem** because there are only two possible outcomes:

- Survived (1)
- Did Not Survive (0)



# Understanding the Titanic Dataset Columns

The Titanic dataset contains information about passengers aboard the RMS Titanic. Each row represents one passenger, while each column represents a specific characteristic (feature) of that passenger. Understanding these columns is important before performing data analysis or building a machine learning model.

---

## 1. PassengerId

**Description:**
PassengerId is a unique identification number assigned to each passenger in the dataset.

**Data Type:** Integer

**Example:**
- 892
- 893
- 894
- 895

**Purpose:**
- Used to uniquely identify each passenger.
- Does not contain information that influences survival.
- Usually removed before training a machine learning model.

---

## 2. Pclass (Passenger Class)

**Description:**
Pclass indicates the travel class or ticket class of the passenger.

**Data Type:** Integer

**Possible Values:**
- 1 = First Class
- 2 = Second Class
- 3 = Third Class

**Examples:**
- 3
- 2
- 1

**Purpose:**
Passenger class is an important feature because passengers in higher classes generally had better access to lifeboats and therefore a higher chance of survival.

---

## 3. Name

**Description:**
The full name of the passenger.

**Data Type:** String (Text)

**Examples:**
- Kelly, Mr. James
- Wilkes, Mrs. James (Ellen Needs)
- Myles, Mr. Thomas Francis

**Purpose:**
Although the full name itself is not usually useful for prediction, titles such as **Mr., Mrs., Miss., Master., Dr.** can provide useful information about gender, age group, or social status.

---

## 4. Sex

**Description:**
Represents the gender of the passenger.

**Data Type:** Categorical

**Possible Values:**
- male
- female

**Examples:**
- male
- female

**Purpose:**
Gender played a major role in survival because women were generally given priority during evacuation.

---

## 5. Age

**Description:**
Represents the age of the passenger in years.

**Data Type:** Numerical (Continuous)

**Examples:**
- 34.5
- 47
- 62
- 27

**Purpose:**
Age helps determine whether younger or older passengers had higher survival rates. Missing values are common and often need to be filled during data cleaning.

---

## 6. SibSp (Siblings/Spouses Aboard)

**Description:**
Indicates the number of siblings or spouses traveling with the passenger aboard the Titanic.

**Data Type:** Integer

**Examples:**
- 0 = Traveling alone
- 1 = One sibling or spouse
- 2 = Two siblings or spouses

**Purpose:**
This feature helps determine whether traveling with close family members affected survival chances.

---

## 7. Parch (Parents/Children Aboard)

**Description:**
Represents the number of parents or children traveling with the passenger.

**Data Type:** Integer

**Examples:**
- 0
- 1
- 2

**Purpose:**
This feature provides information about family size and may influence survival.

---

## 8. Ticket

**Description:**
The ticket number assigned to the passenger.

**Data Type:** String (Text)

**Examples:**
- 330911
- 363272
- 240276
- 315154

**Purpose:**
Ticket numbers may indicate passengers who booked together, but they generally provide little predictive value and are often removed before model training.

---

## 9. Fare

**Description:**
The amount of money paid for the passenger's ticket.

**Data Type:** Numerical (Continuous)

**Examples:**
- 7.8292
- 7.0000
- 9.6875
- 8.6625

**Purpose:**
Passengers who paid higher fares were often in higher classes, which may have increased their chances of survival.

---

## 10. Cabin

**Description:**
The cabin number assigned to the passenger.

**Data Type:** String (Text)

**Examples:**
- C85
- B57 B59 B63 B66
- E46

**Purpose:**
Cabin location may provide useful information about a passenger's location on the ship. However, many cabin values are missing, so this column is often removed or transformed during data preprocessing.

---

## 11. Embarked

**Description:**
The port where the passenger boarded the Titanic.

**Data Type:** Categorical

**Possible Values:**
- C = Cherbourg
- Q = Queenstown
- S = Southampton

**Examples:**
- Q
- S
- C

**Purpose:**
The embarkation port may be associated with passenger class, ticket fare, and survival rates.

---

## 12. Survived (Target Variable)

**Description:**
Indicates whether the passenger survived the Titanic disaster.

**Data Type:** Binary (Categorical)

**Possible Values:**
- 0 = Did Not Survive
- 1 = Survived

**Examples:**
- 0
- 1

**Purpose:**
This is the **target variable** (dependent variable) that we want our machine learning model to predict using all the other features.

---

### 3.2) Visualizing the survied Passangers and those who did not survived using pie chart

```python

# Apply Seaborn style
sns.set_style("whitegrid")

# Count the survival values
survival_counts = data["Survived"].value_counts()

# Create a smaller figure
plt.figure(figsize=(4, 4))

# Create the pie chart
plt.pie(
    survival_counts,
    labels=["Did Not Survive", "Survived"],
    autopct="%1.1f%%",
    startangle=90,
    explode=(0.05, 0.05),
    textprops={"fontsize": 9}
)

# Add title
plt.title("Passenger Survival Distribution", fontsize=10)

# Make the pie chart circular
plt.axis("equal")

# Display the chart
plt.show()
```

## 3.3) Age and Survival
The boxplot compares the age distribution of passengers who survived and those who did not.

```python
plt.figure(figsize=(5,3))

sns.boxplot(
    data=data,
    x="Survived",
    y="Age"
)

plt.title("Age vs Survival")

plt.tight_layout()

plt.show()
```

## 3.4) Survival by Gender

This visualization shows the relationship between a passenger's gender and survival status.

It helps determine whether male and female passengers had different survival rates.

This analysis is important because historical records indicate that women and children were given priority during evacuation.

```python

plt.figure(figsize=(5,3))

sns.countplot(
    data=data,
    x="Sex",
    hue="Survived"
)

plt.title("Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Passengers")

plt.tight_layout()
plt.show()
```

## 3.5) Survival by Passenger Class

This chart illustrates how passenger class influenced survival.

Passengers traveling in higher classes may have had better access to lifeboats compared to passengers in lower classes.

```python
plt.figure(figsize=(5,3))

sns.countplot(
    data=data,
    x="Pclass",
    hue="Survived"
)

plt.title("Survival by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Number of Passengers")

plt.tight_layout()
plt.show()
```

## 3.6) Distribution of Passenger Age

The histogram shows how passenger ages are distributed.

This helps identify whether the dataset contains more children, adults, or elderly passengers.

## 3.9) Correlation Between Numerical Variables

A correlation heatmap shows the strength and direction of relationships between numerical variables.

Correlation values range from:

- +1 (Strong Positive)
- 0 (No Relationship)
- -1 (Strong Negative)

This helps identify variables that are strongly related to survival.

```python
plt.figure(figsize=(6,4))

numeric_data = data.select_dtypes(include=["number"])

sns.heatmap(
    numeric_data.corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap")

plt.tight_layout()

plt.show()
```
